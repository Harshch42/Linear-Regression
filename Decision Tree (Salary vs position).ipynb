# Random Forest Regression

# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing Dataset

```python
dataset=pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

# Training Decision Tree Model on the Whole Dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X,y)
```

# Predicting a new Result

```python
pred_X= regressor.predict(X)
print(pred_X,y)
print(regressor.predict([[6.5]]))
```

# Visualizing the Decision Tree Results High Resolution

```python
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color='red')
plt.title("Organization Employee Salaries (Decision Tree) ")
plt.xlabel('Position of Employee')
plt.ylabel('Salary')
plt.plot(X_grid,regressor.predict(X_grid))
plt.show()
```

Hence , Random Forest Regression is implemented on Salary vs Position dataset
