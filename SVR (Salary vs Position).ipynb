# Support Vector Regression

# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing Dataset

```python
dataset=pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X,y)
```

```python
y=y.reshape(len(y),1)
print(y)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc1 = StandardScaler()
sc2 = StandardScaler()
X = sc1.fit_transform(X)
y = sc2.fit_transform(y)
```

```python
print(X)
```

```python
print(y)
```

# Training SVR Model on the Whole Dataset

```python
from  sklearn.svm import SVR
regressor = SVR(kernel ='rbf')
regressor.fit(X,y)
```

# Predicting a New Result

```python
sc2.inverse_transform(regressor.predict(sc1.transform([[6.5]])).reshape(-1,1))
```

# Visualizing the SVR Results

```python
plt.scatter(sc1.inverse_transform(X),sc2.inverse_transform(y),color='red')
plt.title("Organization Employee Salaries SVR ")
plt.xlabel('Position of Employee')
plt.ylabel('Salary')
plt.plot(sc1.inverse_transform(X),sc2.inverse_transform(regressor.predict(X).reshape(-1,1)))
plt.show()
```

# Visualizing the SVR Results High Resolution

```python
X_grid = np.arange(min(sc1.inverse_transform(X)),max(sc1.inverse_transform(X)),1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(sc1.inverse_transform(X),sc2.inverse_transform(y),color='red')
plt.title("Organization Employee Salaries SVR ")
plt.xlabel('Position of Employee')
plt.ylabel('Salary')
plt.plot(X_grid,sc2.inverse_transform(regressor.predict(sc1.transform(X_grid)).reshape(-1,1)))
plt.show()
```

Hence , SVR is implemented on a Dataset
