# Simple Linear Regression 


Question : The experience in years of some Employees is Given along with their Salaries , train a Regression Model to Find the Salary of a new Employee Based on their Experience.

# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing DataSet

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

# Splitting Data into Training And Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

# Training the Simple Linear Regression Model on the Training Dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

# Predicting The Test Set results

```python
y_pred_test = lr.predict(X_test)
y_pred_train = lr.predict(X_train)
print(lr.predict([[12]]))
```

# Visualizing Training Set results

```python
plt.scatter(X_train,y_train , color='red')
plt.plot(X_train,y_pred_train)
plt.title("Training Set (Salary vs Exp)")
plt.xlabel("Experience (years)")
plt.ylabel("Salary")
plt.show()
```

# Visualizing Test Set results

```python
plt.scatter(X_test,y_test , color='red')
plt.plot(X_train,y_pred_train)
plt.title("Test Set (Salary vs Exp)")
plt.xlabel("Experience (years)")
plt.ylabel("Salary")
plt.show()
```

```python
print(lr.coef_)
print(lr.intercept_)
```
